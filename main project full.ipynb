### <span style="color:#b30000;">2. Removing Duplicates</span>

Duplicate rows can occur during data collection or merging.  
They can distort the analysis and statistical summaries, so it’s important to remove them.  
We’ll check for duplicate records and remove them if any exist.


```python
# Check for duplicate rows
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows before removal: {duplicate_count}")

# Remove duplicates
df.drop_duplicates(inplace=True)

# Check again after removal
duplicate_count_after = df.duplicated().sum()
print(f"Number of duplicate rows after removal: {duplicate_count_after}")
```

### <span style="color:#b30000;">3. Mapping Coded Columns to Readable Labels</span>



```python
# Map Numeric Codes to Meaningful Labels

# Gender
df['Gender'] = df['RIAGENDR'].map({1.0: 'Male', 2.0: 'Female'}).fillna('Unknown')

# Diabetes Status
df['Diabetes_Status'] = df['DIQ010'].map({1.0: 'Yes', 2.0: 'No', 3.0: 'Unknown'}).fillna('Unknown')

# Prediabetes Status
df['Prediabetes_Status'] = df['DIQ050'].map({1.0: 'Yes', 2.0: 'No', 3.0: 'Unknown'}).fillna('Unknown')

# Heart Disease
df['Heart_Disease'] = df['BPQ020'].map({1.0: 'Yes', 2.0: 'No'}).fillna('Unknown')

# Blood Pressure Medication
df['BP_Medication'] = df['BPQ050A'].map({1.0: 'Yes', 2.0: 'No'}).fillna('Unknown')

# Smoking Status
df['Smoking_Status'] = df['CDQ010'].map({1.0: 'Yes', 2.0: 'No'}).fillna('Unknown')

# Passive Smoke Exposure (Secondhand Smoke)
df['Passive_Smoke_Exposure'] = df['CDQ001'].map({1.0: 'Yes', 2.0: 'No'}).fillna('Unknown')

# Alcohol Consumption
df['Alcohol_Consumption'] = df['DBQ700'].map({1.0: 'Yes', 2.0: 'No'}).fillna('Unknown')

# Blood Pressure Status
df['BP_Status'] = df['BPQ060'].map({1.0: 'High', 2.0: 'Normal'}).fillna('Unknown')

# High Cholesterol
df['High_Cholesterol'] = df['BPQ070'].map({1.0: 'Yes', 2.0: 'No'}).fillna('Unknown')

# High Blood Sugar
df['High_Blood_Sugar'] = df['BPQ090D'].map({1.0: 'Yes', 2.0: 'No'}).fillna('Unknown')

# Diet Quality
def map_diet_quality(score):
    if pd.isna(score):
        return 'Unknown'
    elif score <= 2:
        return 'Poor'
    elif score <= 4:
        return 'Fair'
    else:
        return 'Good'

df['Diet_Quality'] = df['CDQ001'].apply(map_diet_quality)
```

### <span style="color:#b30000;">5. Creating BMI Category Column</span>

Body Mass Index (BMI) is a key health indicator used to classify individuals into  
categories such as *Underweight, Normal, Overweight,* and *Obese.*  

We will create a new column **`BMI_Category`** using standard WHO classification rules:
- BMI < 18.5 → Underweight  
- BMI 18.5 – 24.9 → Normal  
- BMI 25.0 – 29.9 → Overweight  
- BMI ≥ 30 → Obese

```python
# Create a new column 'BMI_Category' based on BMI ranges
def bmi_category(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Normal'
    elif 25 <= bmi < 30:
        return 'Overweight'
    elif bmi >= 30:
        return 'Obese'
    else:
        return 'Unknown'

df['BMI_Category'] = df['BMXBMI'].apply(bmi_category)

# Display sample output
df[['RIDAGEYR', 'Gender', 'BMXBMI', 'BMI_Category']].head(10)
```

<h3 style="color:BLUE;">8. Heart Disease by Gender</h3> 

```python
# Filter dataset to remove 'Unknown'
df_gender = df[df['Heart_Disease'] != 'Unknown']

# Create grouped data by gender and heart disease
heart_by_gender = df_gender.groupby('Gender')['Heart_Disease'].value_counts(normalize=True).unstack() * 100

# Plot side-by-side pies
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

colors = ['#5DADE2', '#E67E22']  # blue and orange

# Pie 1: Heart Disease = Yes
axes[0].pie(
    heart_by_gender['Yes'],
    labels=heart_by_gender.index,
    autopct='%1.1f%%',
    colors=colors,
    startangle=90,
    textprops={'fontsize': 10}
)
axes[0].set_title('Heart Disease: Yes', fontweight='bold')

# Pie 2: Heart Disease = No
axes[1].pie(
    heart_by_gender['No'],
    labels=heart_by_gender.index,
    autopct='%1.1f%%',
    colors=colors,
    startangle=90,
    textprops={'fontsize': 10}
)
axes[1].set_title('Heart Disease: No', fontweight='bold')

plt.suptitle('Heart Disease by Gender', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

<h2 style="color:#003366;"><b>Insight</b></h2>

>Among participants with heart disease, females slightly outnumber males (52.1% vs. 47.9%).

>This suggests both genders face similar heart risk, with a mildly higher rate in females.

>Possible influencing factors: age, lifestyle, and hormonal balance.

>Health awareness and preventive programs should target both men and women equally.
